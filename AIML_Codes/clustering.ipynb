```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
ds = load_iris()
ds
```

```python
ds.feature_names
ds.data
ds.target_names
ds.target
```

```python
df= pd.DataFrame(ds.data, columns=ds.feature_names)
df
```

```python
df['target'] = ds.target
df
```

```python
X=df.iloc[:,2:4]
X
```

```python
#dendogram
import scipy.cluster.hierarchy as ch
link = ch.linkage(X, method = 'ward')
ch.dendrogram(link)
```

AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
ac
```

```python
y_cluster = ac.fit_predict(X)
y_cluster
labels = ac.labels_
labels
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', data = X, hue = labels, palette = 'rainbow')
```

KMean

```python
from sklearn.cluster import KMeans
a=[]
for i in range(1,8):
  km = KMeans(n_clusters=i)
  y_pred = km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
  a.append(km.inertia_ )
print(a)
plt.plot(range(1,8),a)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
sns.scatterplot(x = 'sepal length (cm)', y = 'sepal width (cm)', data = X, hue = labels, palette = 'rainbow')
```
